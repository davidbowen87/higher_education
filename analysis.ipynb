```python
# import dependencies 
import pandas as pd
```

```python
# Load the data
highered_data = pd.read_csv("Resources/IS_HR_2021.csv")
```

```python
# Review the data 
highered_data.head()
```

```python
# check data types
highered_data.dtypes
```

```python
# change data types for the columns that we need
highered_data = highered_data.astype({"FACSTAT": str, "ARANK": str, "CONTROL": str, "HLOFFER": str, "INSTSIZE": str})
highered_data.dtypes
```

```python
df1 = highered_data.loc[0:,["UNITID", "HRTOTLT", "HRTOTLM", "HRTOTLW", "CONTROL", "HLOFFER", "INSTSIZE"]].drop_duplicates()
df1.head()
```

```python
# Get the faculty totals for each institution

df2 = df1.groupby(["UNITID"], as_index = False).agg({"HRTOTLT": "sum", "HRTOTLM":"sum", "HRTOTLW":"sum", "CONTROL":"first", "HLOFFER":"first", "INSTSIZE":"first"})
df2.head()
```

```python
df2.describe()
```

```python
# Calculate the number of institutions
institution_total = df2["UNITID"].count()
institution_total
```

```python
# statistics by control type
control1_df = df2.loc[df2["CONTROL"] =='1']
control1_df
```

```python
control2_df = df2.loc[df2["CONTROL"] =='2']
control2_df
```

```python
control3_df = df2.loc[df2["CONTROL"] =='3']
control3_df
```

```python
# Calculate the number of public schools
public_total = control1_df["UNITID"].count()
public_total
```

```python
# Calculate the number of not for profit schools
not_for_profit_total = control2_df["UNITID"].count()
not_for_profit_total
```

```python
# Calculate the number of private for profit schools
private_for_profit_total = control3_df["UNITID"].count()
private_for_profit_total
```

```python
# Calculate the percentage for each control class
public_percentage = (public_total / institution_total)*100
not_for_profit_percentage = (not_for_profit_total / institution_total)*100
private_for_profit_percentage = (private_for_profit_total / institution_total)*100

public_percentage
```

```python
# Stats by faculty's academic rank
df3 = highered_data.loc[0:,["UNITID", "ARANK", "HRTOTLT", "HRTOTLM", "HRTOTLW"]]
df3.head()
```

```python
# totals by academic rank

arank_df = df3.groupby(["ARANK"], as_index = False).agg({"HRTOTLT": "sum", "HRTOTLM":"sum", "HRTOTLW":"sum"})
arank_df
```
