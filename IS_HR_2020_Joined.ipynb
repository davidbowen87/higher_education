Finally, we will save the joined data to a new CSV file named IS_HR_2020.csv 

```python
# import dependencies 
import pandas as pd
```

```python
# Load the data
hr_data = pd.read_csv("Resources/HR2020_reduced.csv")
is_data = pd.read_csv("Resources/IS2020_reduced.csv")
```

```python
# Review the data
hr_data.head()
```

```python
# Review the data 
is_data.head()
```

```python
# Join the data
joined_data = pd.merge(hr_data, is_data, on="UNITID")
joined_data.head()
```

```python
# Review data types of columns 
joined_data.info()
```

```python
# Summary statistics for the numeric columns in the DataFrame
print(joined_data.describe())
```

```python
# Returns the number of missing values for each column in the DataFrame
print(joined_data.isnull().sum())
```

```python
# Returns the number of duplicated rows in the DataFrame
print(joined_data.duplicated().sum())
```

```python
# Change the column name
joined_data = joined_data.rename(columns={"Year_x": "Year"})
```

```python
# Drop the Year_y column
joined_data = joined_data.drop(columns=["Year_y"])
```

```python
# Review data types of columns 
joined_data.info()
```

```python
# check the dataframe
joined_data.head()
```

```python
# Save the joined data
joined_data.to_csv("Resources/IS_HR_2020.csv", index=False)
```
